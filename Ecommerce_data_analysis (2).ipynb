# Convert date column

```python
orders['order_date'] = pd.to_datetime(orders['order_date'])
reviews['review_date'] = pd.to_datetime(reviews['review_date'])
events['event_timestamp'] = pd.to_datetime(events['event_timestamp'])
```

# Adding new column like month,year and dayofweek

```python
orders['Order_Month'] = orders['order_date'].dt.month 
orders['Order_Year'] = orders['order_date'].dt.year
orders['Order_Day_of_Week'] = orders['order_date'].dt.dayofweek
reviews['Order_Month'] = reviews['review_date'].dt.month 
reviews['Order_Year'] = reviews['review_date'].dt.year
reviews['Order_Day_of_Week'] = reviews['review_date'].dt.dayofweek
events['Order_Month'] = events['event_timestamp'].dt.month 
events['Order_Year'] = events['event_timestamp'].dt.year
events['Order_Day_of_Week'] =events['event_timestamp'].dt.dayofweek
```

# Total Revenue using order_items table

```python
total_revenue = order_items["item_total"].sum()

print("Total Revenue:", total_revenue)
```

# Average amount spent per order.

```python
avg_order_value = orders["total_amount"].mean()

print("Average Order Value:", avg_order_value)
```

# Order Status Distribution

```python

status_counts = orders["order_status"].value_counts()

print(status_counts)

status_counts.plot(kind="bar")
plt.title("Order Status Distribution")
plt.show()
```

# Monthly Order Trend

```python
orders["order_date"] = pd.to_datetime(orders["order_date"])

monthly_orders = orders.groupby(
    orders["order_date"].dt.to_period("M")
).size()

monthly_orders.plot()
plt.title("Monthly Order Trend")
plt.ylabel("Orders")
plt.show()
```

# Top Selling Products

```python
df = pd.merge(order_items, products, on="product_id")

top_products = df.groupby("product_name")["quantity"].sum().sort_values(ascending=False).head(10)

print(top_products)
```

```python
top_products.plot(kind="bar")
plt.title("Top Selling Products")
plt.ylabel("Units Sold")
plt.show()
```

# Revenue by Category

```python
category_revenue = df.groupby("category")["item_total"].sum().sort_values(ascending=False)

print(category_revenue)
```

```python
category_revenue.plot(kind="bar")
plt.title("Revenue by Category")
plt.show()
```

# Customer Purchase Frequency

```python
purchase_freq = orders.groupby("user_id")["order_id"].count()

print(purchase_freq.describe())
```

```python
purchase_freq.hist(bins=20)
plt.title("Customer Purchase Frequency")
plt.show()
```

# Top Customers by Spending

```python
customer_spending = orders.groupby("user_id")["total_amount"].sum().sort_values(ascending=False).head(10)

print(customer_spending)
```

# Funnel Analysis (Views â Cart)

```python
funnel = events["event_type"].value_counts()

print(funnel)
```

```python
funnel.plot(kind="bar")
plt.title("User Funnel Events")
plt.show()
```

# User Engagement (Active vs inactive users)

```python
user_activity = events.groupby("user_id")["event_type"].count()

user_activity.hist(bins=30)
plt.title("User Engagement Distribution")
plt.show()
```

# Rating Distribution (Customer satisfaction levels)

```python
rating_counts = reviews["rating"].value_counts().sort_index()

print(rating_counts)
```

# Average Rating per Product

```python
avg_rating = reviews.groupby("product_id")["rating"].mean().sort_values(ascending=False).head(10)

print(avg_rating)
```

# Revenue by Brand (Brands generating most revenue)

```python
brand_revenue = df.groupby("brand")["item_total"].sum().sort_values(ascending=False)

print(brand_revenue.head(10))
```

# Customer Segmentation (A small group of high-value customers contributes significantly to revenue)

```python
customer_spending = orders.groupby("user_id")["total_amount"].sum()

segments = pd.qcut(customer_spending, q=3, labels=["Low Value", "Medium Value", "High Value"])

segment_counts = segments.value_counts()

segment_counts.plot(kind="bar")
plt.title("Customer Segmentation")
plt.show()
```

# City-wise Customer Distribution

```python
city_users = users["city"].value_counts().head(10)

city_users.plot(kind="bar")
plt.title("Top Cities by Number of Users")
plt.show()
```

# Analyze Cancelled Orders

```python
cancel_rate = orders["order_status"].value_counts(normalize=True) * 100

print(cancel_rate)
```

# Product Rating vs Sales Analysis(Higher-rated products tend to sell more)

```python
sales_rating = df.groupby("product_id").agg({
    "quantity":"sum",
    "rating":"mean"
})

print(sales_rating.head())
```
